```python
import pandas as pd
import numpy as np
```

```python
invoice_csv = pd.read_csv('Invoice_20201220.csv', sep= '|')
```

```python
invoice_csv.head()
```

```python
invoice_csv.drop(columns='INVOICE_NAME', inplace = True)
```

```python
invoice_csv['SITE_CODE']=invoice_csv['SITE_NAME'].str[-7:-1]
invoice_csv.head()
```

```python
import datetime as dt

invoice_csv['NB_DAYS']= ((pd.to_datetime(invoice_csv['END'], format="%d/%m/%Y")) - (pd.to_datetime(invoice_csv['BEGIN'], format="%d/%m/%Y"))).dt.days + 1
invoice_csv.head()
```

```python
import re
# j'ai testÃ© beaucoup de choses pour remplacer le KWH str par un integer, apparemment y en a une qui a marchÃ© pq c'est maintenant un integer.. mais je ne sais pas laquelle ðŸ˜… peut Ãªtre pd.to_numeric avec un replace de l'espace dans le string de dÃ©part
invoice_csv['DAILY_CONS']=round(invoice_csv['KWH_PERIOD']/invoice_csv['NB_DAYS'],2)
invoice_csv.head()
```

```python
invoice_csv['NB_INVOICES']=invoice_csv.groupby(['SITE_CODE'])['SITE_CODE'].transform('count')
invoice_csv
```

```python
coef_variation = (invoice_csv.groupby(["SITE_CODE"])["DAILY_CONS"].std() / invoice_csv.groupby(["SITE_CODE"])["DAILY_CONS"].mean()).map('{:.2%}'.format)
print(coef_variation)
```

```python
invoice_csv=invoice_csv[['SITE_CODE','BEGIN','END','KWH_PERIOD','PERIOD_EUR','NB_DAYS','DAILY_CONS','NB_INVOICES']]
invoice_csv.head()
```

```python
invoice_csv.to_csv('invoices_clean.csv')
```
